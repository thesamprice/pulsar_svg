```python
import os
import sys
def PyGet_filenames():
    folder = 'ascii/'
    files = os.listdir(folder)
    
    
    items = []
    for filename in files:
        
        filename = './ascii/' + filename
        
    
    
        data = ReadD3(filename)
    
        
        lo,hi = sys.maxint,-sys.maxint-1
        for x in (item['mag'] for item in data):
            lo,hi = min(x,lo),max(x,hi)
        items.append({'filename':filename,'lo':lo,'hi':hi})
    
    newlist = sorted(items, key=lambda k: k['hi']) 
    newlist.reverse()
    return newlist

print PyGet_filenames()
```

```python
def ReadASC(filename):

    fid = open(filename)
    data = fid.read()
    fid.close()

    import re
    regex = 'EXTNAME\s*=\s*(\w+)'
    import numpy
    groups = []
    for a in re.finditer(regex,data):
        name = a.group(1)
        p_data ={ 'start':a.span()[0]}
        groups.append(p_data)
    
    for i in range(len(groups)-1):
        groups[i]['end'] = groups[i+1]['start']
        
    groups[-1]['end'] = len(data)
    all_data = {}
    for a in groups:
        s = data[a['start']:a['end']]
        
        regex = '\s*([a-zA-Z0-9_]+)\s*=\s*([a-zA-Z0-9_]+)'
        for pair in re.finditer(regex,s):
            d = pair.groups()
            a[d[0]] = d[1]
            
        start = re.search('\n\s+1\s+',s).span()[0]
        s = s[start:].strip()
        a['data'] = []
        a['NCOLS'] = int(a['NCOLS'])
        a['NROWS'] = int(a['NROWS'])
        array = numpy.zeros((a['NCOLS']+1,a['NROWS']))
        #array = numpy.zeros((a['NROWS'],a['NCOLS']))    
    
        #print array
        #print s
        row_count=0    
        for line in s.splitlines():
            
            d = []
            col_count = 0
            for col in line.split():
                #print row_count,col_count
                #array[row_count][col_count] = float(col)
                array[col_count][row_count] = float(col)
                
                col_count += 1
            row_count +=1
            #print array
                
        a['data'] = array.tolist()
        all_data[a['EXTNAME'] ]= a
    
    return all_data
def ReadD3(filename):
    import numpy as np
    data = ReadASC(filename)
    parts = range(len(data['BEST_FIT_LC']['data'][0]));
    p_data = data['BEST_FIT_LC']['data'][3]
    max_ind = np.argmax(p_data)
    #max_val = np.max(p_data)
    import random
    p_data = np.roll(p_data,-max_ind + len(p_data)/2 + random.randint(-60,60))
    
    result = []
    for ind in parts:
        phase = data['BEST_FIT_LC']['data'][1][ind] #Phase start, to #Phase end is row 2
        mag   = data['BEST_FIT_LC']['data'][3][ind]
        mag = p_data[ind]
        row = {'phase':phase,'mag':mag/1.5}
        result.append(row)
    return result
        
#ReadD3('/Users/sprice/src/pulsar_poster/2PC_auxiliary_files_v04/PSR_data/ascii/PSRJ0034-0534_2PC_data.asc')
```
